```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 值，从 0 到 2π，共 100 个点
x = np.linspace(0, 2 * np.pi, 100)

# 计算对应的余弦值
y = np.cos(x)

# 创建图形
plt.figure(figsize=(8, 4))

# 绘制余弦函数
plt.plot(x, y, label='cos(x)')

# 添加标题和坐标轴标签
plt.title('余弦函数')
plt.xlabel('x (弧度)')
plt.ylabel('cos(x)')

# 显示图例
plt.legend()

# 显示网格
plt.grid(True)

# 显示图形
plt.show()
```
